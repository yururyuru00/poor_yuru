```python
import seaborn as sns
import math
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

debug = "test"
useMat = 'w'
triNumber = 1
blockNumber = 1
for index in [0,1,2,3,4]:
    if (useMat == 'sp'):
        path = 'C:/Users/yajima/Java/ResearchSNSClustering/' + debug + '/sp/' + useMat + str(index) + '.txt'
    else:
        path = 'C:/Users/yajima/Java/ResearchSNSClustering/' + debug + '/tri_' + str(triNumber) + '/block_' + str(blockNumber) + '/' + useMat + str(index) + '.txt'
    with open(path, 'r') as f:
        size = int(f.readline())
        ary = np.zeros((size, size))
        i=0
        for line in f.readlines():
            j=0
            lines = line.rstrip().split(" ")
            for val in lines:
                ary[i][j] = float(val)
                j += 1
            i += 1
    plt.figure(figsize=(36, 36))
    sns.heatmap(ary, fmt='g', cmap='gray_r')
    
    
```
